```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,6))
x["TotalGHGEmissions"]=(x["TotalGHGEmissions"]-x["TotalGHGEmissions"].mean())/x["TotalGHGEmissions"].std()
sns.boxplot(x='BuildingType',y="TotalGHGEmissions", data=x)
```

```python
sns.catplot(x='BuildingType',kind='count',height=6,aspect=2,data=x)
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=x['Electricity(kBtu)'],y=x['GHGEmissions(MetricTonsCO2e)'])
```

```python
x['GHGEmissionsIntensity(kgCO2e/ft2)']=(x['GHGEmissionsIntensity(kgCO2e/ft2)'])/x['GHGEmissionsIntensity(kgCO2e/ft2)'].max()
plt.figure(figsize=(15,8))
sns.lineplot(x='BuildingType',y='GHGEmissionsIntensity(kgCO2e/ft2)',data=x)
```
